```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
data=pd.read_csv("Fat_Supply_Quantity_Data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.fillna(method="ffill",inplace=True)
```

```python
data.Undernourished=data.Undernourished.str.replace("<","")
data.Undernourished=data.Undernourished.astype('float')
```

```python
data
```

```python
x=data[["Obesity","Undernourished"]]
y=data.Recovered
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1)
x_test
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,x_vars=["Obesity","Undernourished"],y_vars=["Recovered"],diag_kind=None)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#statsmodels
import statsmodels.formula.api as sm
```

```python
lr=sm.ols(formula='Recovered ~ Obesity + Undernourished',data=data).fit()
```

```python
y_sm_predict=lr.predict({"Obesity":28.4,"Undernourished":2.5 })
y_sm_predict
```
