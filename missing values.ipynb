```python
data.interpolate(method="linear")
```

```python
from sklearn.impute import SimpleImputer
```

```python
cols=data.loc[:,["Obesity","Undernourished","Confirmed","Deaths","Recovered","Active"]]
cols
```

```python
my_imputer=SimpleImputer(strategy="most_frequent")
```

```python
imputed_values=my_imputer.fit_transform(cols)
imputed_values
```

```python
new_values=pd.DataFrame(imputed_values)
new_values
```

```python
new_values.columns=cols.columns
new_values
```

```python
data["Obesity"]=new_values["Obesity"]
data["Undernourished"]=new_values["Undernourished"]
data["Confirmed"]=new_values["Confirmed"]
data["Deaths"]=new_values["Deaths"]
data["Recovered"]=new_values["Recovered"]
data["Active"]=new_values["Active"]
```
