```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset("iris")
data
```

```python
plt.scatter(data.petal_length,data.petal_width)
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.show()
```

```python
cols=data.loc[:,["petal_length","petal_width"]]
cols
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
cluster_labels=km.fit_predict(cols)
cluster_labels
```

```python
data["culster_labels"]=cluster_labels
```

```python
data
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
km.n_iter_
```

```python
group0=data[data.culster_labels==0]
group0
```

```python
group1=data[data.culster_labels==1]
group2=data[data.culster_labels==2]
```

```python
plt.scatter(group0.sepal_length,group0.sepal_width,label="cluster = 0")
plt.scatter(group1.sepal_length,group1.sepal_width,label="cluster = 1")
plt.scatter(group2.sepal_length,group2.sepal_width,label="cluster = 2")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],label="Centeriods ",marker="*",color="black")
plt.legend()
plt.show()
```
